```python
import plotly.io as pio
pio.renderers.default = 'notebook'  # or 'notebook_connected', 'iframe', or 'vscode' depending on your setup
```

```python
import plotly.express as px

fig = px.scatter(
    tls_qc,
    x='tls_overlap_percentage',
    y='qc_overlap_percentage',
    hover_name='sample_name',
    hover_data=['qc_type', 'tls_id'],
    title='TLS vs QC Overlap Percentage',
    labels={
        'tls_overlap_percentage': 'TLS Overlap (%)',
        'qc_overlap_percentage': 'QC Overlap (%)',
        'qc_type': 'QC Type',
        'tls_id': 'TLS ID'
    }
)

fig.update_traces(marker=dict(size=8, line=dict(width=1, color='DarkSlateGrey')))
fig.update_layout(height=600, width=800)

fig.show()
```
