# JSON File Ratios

```python
import json
import os
```

```python
with open('TCGA-IB-AAUU-01Z-00-DX1.83137319-8229-4682-8BD3-9B9A5C6C997A.geojson', 'r') as file:
    json_file = json.load(file)

# Print the data
print(json_file)
```

```python
for feature in json_file['features']:
    for polygon in feature['geometry']['coordinates']:
        for coord in polygon:
            for i in range(len(coord)):
                coord[i] *= 2
```

```python
with open('modified_TCGA-IB-AAUU-01Z-00-DX1.83137319-8229-4682-8BD3-9B9A5C6C997A.geojson', 'w') as outfile:
    json.dump(json_file, outfile, indent=4)
```
